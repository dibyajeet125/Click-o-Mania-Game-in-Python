```python
# Import necessary libraries
import tkinter as tk
import random
from functools import partial

# Create the main window
window = tk.Tk()
window.geometry("500x500")
window.title('Clickomania game')

# Create three frames for navigation, main content, and footer
Nav = tk.Frame(master=window, pady=5, padx=16)
Main = tk.Frame(master=window, pady=16, padx=16)
Footer = tk.Frame(master=window, pady=16, padx=16)
Nav.pack(expand=True)
Main.pack(expand=True)
Footer.pack(expand=True)

# Initialize variables
Buttons = []  # List to store button widgets
Score = 25    # Initial score
Match_btn = []  # List to store buttons that are selected for comparison
Guesses = 0   # Counter for the number of guesses

# Create a label to display the score
Label_Score = tk.Label(master=Nav, text=f"Score: {Score}", font=("Times new roman", 30))
Label_Score.pack()

# Create a "Reset" button
Reset_btn = tk.Button(master=Footer, text="Reset", font=("Times new roman", 30), padx=10, pady=3)
Reset_btn.pack()

# Function to handle button click events
def Activate(btn: object, color: str):
    global Match_btn
    global Score
    global Match_btn
    global Label_Score
    global Label_Lost
    global Guesses

    # Check if the button background matches the color
    if btn['background'] == color:
        btn['background'] = 'grey'
        Match_btn.pop()
    else:
        btn['background'] = color
        Match_btn.append(btn)

    # Check if two buttons are selected for comparison
    if len(Match_btn) == 2:
        if Match_btn[0]['background'] == Match_btn[1]['background']:
            # If the colors match, disable the buttons and update the score
            Match_btn[0].config(command='')
            Match_btn[1].config(command='')
            Match_btn[0]['text'] = color
            Match_btn[1]['text'] = color
            Guesses += 1
        else:
            # If the colors don't match, reset the button backgrounds and decrease the score
            Match_btn[0]['background'] = 'grey'
            Match_btn[1]['background'] = 'grey'
            Score -= 1
            Score_Update(Score)

            # Check if the player has lost (score reaches 0)
            if Score == 0:
                for btn in Buttons:
                    btn.destroy()

                Label_Lost.pack(side=tk.TOP, expand=True, fill=tk.BOTH)

        # Check if the player has won (reached the maximum number of guesses)
        if Guesses == 8:
            for btn in Buttons:
                btn.destroy()

            Label_Won.pack(side=tk.TOP, expand=True, fill=tk.BOTH)

        # Clear the list of buttons selected for comparison
        Match_btn.clear()

# Function to initialize the game
def Game():
    global Score
    global Guesses
    global Label_Score
    global Buttons
    global Match_btn
    global Label_Lost
    global Label_Won
    global Nav

    # Reset variables
    Match_btn = []
    Guesses = 0
    Score = 25
    Colors = ['white', 'black', 'red', 'green','blue', 'cyan', 'yellow', 'magenta'] * 2

    # Update the score display
    Score_Update(Score)

    # Destroy previous labels and buttons
    if len(Buttons):
        Label_Won.destroy()
        Label_Lost.destroy()
        for btn in Buttons:
            btn.destroy()

    # Create labels for winning and losing
    Label_Won = tk.Label(master=Main, text="You WON!", font=("Times new roman", 35))
    Label_Lost = tk.Label(master=Main, text="You Lost, try again!", font=("Times new roman", 35),)

    # Create buttons with random colors for the game grid
    for i in range(4):
        for j in range(4):
            Random_Color = random.randint(0, len(Colors)-1)
            color = Colors[Random_Color]

            Square_btn = tk.Button(master=Main, text=f"????", width=10, height=3, background='grey', activebackground=color)
            Buttons.append(Square_btn)
            Square_btn.grid(row=i, column=j, padx=10, pady=10)
            Buttons[-1]['command'] = partial(Activate, Square_btn, color)

            Colors.pop(Random_Color)

# Function to update the score display
def Score_Update(Score: int):
    Label_Score['text'] = f"Score: {Score}"

# Set the command for the Reset button to start a new game
Reset_btn['command'] = Game

# Start the initial game
Game()

# Start the Tkinter main loop
window.mainloop()
```
